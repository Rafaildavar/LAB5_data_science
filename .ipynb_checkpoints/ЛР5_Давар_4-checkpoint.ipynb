```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, balanced_accuracy_score, confusion_matrix, roc_curve, roc_auc_score
```

Набор данных **4heart2.csv**

Данные о болезнях сердца:
 1. возраст: возраст пациента (лет)
 2. анемия: снижение количества эритроцитов или гемоглобина (логическое значение)
 3. высокое кровяное давление: если у пациента гипертония (логическое значение)
 4. креатининфосфокиназа КФК уровень фермента КФК в крови (мкг/л)
 5. диабет: если у пациента диабет (логическое значение)
 6. фракция выброса: процент крови, покидающей сердце при каждом сокращении (в процентах)
 7. тромбоциты: тромбоциты в крови (килотромбоциты/ мл)
 8. пол: женщина или мужчина (бинарный)
 9. креатинин сыворотки: уровень креатинина сыворотки в крови (мг/дл)
 10. натрий сыворотки: уровень натрия сыворотки в крови (мэкв/л)
 11. курение: если пациент курит или нет (логическое)
 12. время: период наблюдения (дни)
 13. событие смерти: если пациент умер в течение периода наблюдения (логическое значение)

Для начала необходимо загрузить датасет, после чего проводится обработка данных для дальнейшей работы с загруженным датасетом.

```python
df = pd.read_csv('4heart2.csv', sep=",")
df.head()
```

```python
df.info()
```

Датасет содержит 299 записей, а пропусков нет. Все данные количественные.

Далее данные проверяются на наличие явных дубликатов с помощью метода duplicated().

```python
display(df.duplicated().sum())
```

Теперь производится поиск неявных дубликатов или ошибочных данных, для этого просматриваются уникальные значения с помощью метода unique().

```python
print(df['age'].unique())
print(df['anaemia'].unique())
print(df['high_blood_pressure'].unique())
print(df['diabetes'].unique())
print(df['sex'].unique())
print(df['smoking'].unique())
print(df['time'].unique())
print(df['DEATH_EVENT'].unique())
print(df['ejection_fraction'].unique())
```

В датасете нет ошибочных данных или неявных дубликатов.

Целевой переменной был выбран DEATH_EVENT. Чтобы выявить что чаще всего приводит к гибели была построена матрица диаграмм рассеяния, где эта переменная выделена цветом.

```python
df = df.rename(columns={'DEATH_EVENT': 'Target'})
```

```python
plt.figure(figsize=(12, 6))
sns.pairplot(df, hue='Target')
plt.show()
```

Закономерности из графиков выявить тяжело.

Целевая переменная должна быть исключена из модели, выделена в y. Разделим данные на обучающую и тестовую выборки, стандартизируем числовые данные с помощью StandartScaler.

```python
y = df['Target']
df.drop(columns=['Target'], inplace=True)
x = df

x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.25, random_state=20)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_valid_scaled = scaler.transform(x_valid)
```

Были разработаны следующие модели:
- метод k- ближайших соседей
- логистическая регрессия
- дерево решений
- случайный лес  

Для каждой модели воспользуемся GridSearchCV, чтобы найти лучшие значения гиппер-параметров


### k-ближайших соседей

Используя GridSearchCV, переберём все комбинации параметров модели к-ближайших, чтобы найти лучшие значения гиперпараметров.

```python
parametrs = {
    "n_neighbors": [2, 3, 4, 5, 8, 10],
    "algorithm": ['ball_tree', 'kd_tree', 'brute'],
    "leaf_size": [5, 10, 20, 30, 40, 50]
    }

score_metrics = ['neg_mean_absolute_error', 'neg_mean_squared_error', 'r2', 'neg_root_mean_squared_error']

model_GC = KNeighborsClassifier()
grid = GridSearchCV(model_GC,
                    parametrs,
                    scoring='recall',
                    cv=6)

grid.fit(x_train, y_train)
grid.best_params_
```

Была обучена модель, используя найденные гиперпараметры и выведены её метрики.

```python
neighbors_model = KNeighborsClassifier(n_neighbors=3, algorithm='ball_tree', leaf_size=5)
neighbors_model.fit(x_train, y_train)

y_pred = neighbors_model.predict(x_valid)

acc = accuracy_score(y_valid, y_pred)
precision = precision_score(y_valid, y_pred)
recall = recall_score(y_valid, y_pred)
f1 = f1_score(y_valid, y_pred)
balanced_accuracy = balanced_accuracy_score(y_valid, y_pred)
print(classification_report(y_valid, y_pred))
print('balanced accuracy: ', balanced_accuracy)
print('accuracy: ', acc)
print('precision: ', precision)
print('recall: ', recall)
print('f1: ', f1)
```

* Balanced accuracy — это среднее значение точности для каждого класса, учитывая дисбаланс классов. 0.527 указывает на то, что модель имеет несколько лучшую производительность по сравнению с случайным выбором, но всё ещё далека от идеала.
* Accuracy — это доля правильных прогнозов относительно всех сделанных прогнозов. Точность 0.613 говорит о том, что модель в целом предсказала правильно 61.3% случаев.
* Precision показывает, насколько из предсказанных положительных случаев действительно были положительными. 33.3% из всех предсказанных положительных случаев на самом деле таковыми и являются. Это низкий показатель, что говорит о том, что модель часто ошибается, классифицируя отрицательные примеры как положительные (false positives).
* Recall — это способность модели правильно находить все положительные примеры, то есть доля истинных положительных среди всех положительных случаев. 31.8% указывает на то, что модель не может найти большинство положительных примеров.
Это низкий показатель, что означает, что модель не находит достаточно положительных примеров, возможно, из-за ошибки в настройке гиперпараметров или из-за сильного дисбаланса классов.
* F1-Score — это гармоническое среднее между точностью и полнотой. Он используется для оценки модели в случае, если важно учитывать как точность, так и полноту. 0.326 указывает на то, что модель имеет довольно слабую производительность, так как F1-Score ниже 0.5 считается плохим результатом.
Это ещё одно подтверждение того, что модель не справляется с задачей в её текущем виде

### Дерево решений

Используя GridSearchCV, переберём все комбинации параметров модели дерева решений, чтобы найти лучшие значения гиперпараметров.

```python
parametrs = {
    "criterion": ['gini', 'entropy', 'log_loss'],
    "splitter": ['best', 'random'],
    "max_depth": [2, 3, 4, 5, 6, 7, 8],
    "random_state": [20]
    }

model_GC = DecisionTreeClassifier()
grid = GridSearchCV(model_GC,
                    parametrs,
                    scoring='recall',
                    cv=6)

grid.fit(x_train, y_train)
grid.best_params_
```

Была обучена модель, используя найденные гиперпараметры и выведены её метрики.

```python
tree_model = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=5, random_state=20)
tree_model.fit(x_train, y_train)

y_pred = tree_model.predict(x_valid)

acc = accuracy_score(y_valid, y_pred)
precision = precision_score(y_valid, y_pred)
recall = recall_score(y_valid, y_pred)
f1 = f1_score(y_valid, y_pred)
balanced_accuracy = balanced_accuracy_score(y_valid, y_pred)
print(classification_report(y_valid, y_pred))
print('balanced accuracy: ', balanced_accuracy)
print('accuracy: ', acc)
print('precision: ', precision)
print('recall: ', recall)
print('f1: ', f1)
```

* Balanced accuracy = 0.752 — это хороший результат, что указывает на то, что модель неплохо справляется с обработкой дисбаланса классов и демонстрирует хорошие результаты по сравнению с случайным выбором.
* Accuracy = 0.8 — отличный результат, который говорит о том, что модель в 80% случаев правильно классифицирует объекты. Это довольно высокий показатель, и он говорит о хорошей общей производительности модели.
* Precision = 66.7%. Это хороший результат, особенно если важно избежать ложных положительных классификаций.
* Recall = 63.6%. Это довольно неплохой показатель, означающий, что модель достаточно хорошо находит положительные примеры.
* F1-Score = 0.651 говорит о том, что модель довольно хорошо балансирует между точностью и полнотой.


### Логистическая регрессия

Используя GridSearchCV, переберём все комбинации параметров модели логистической регрессии, чтобы найти лучшие значения гиперпараметров.

```python
parametrs = {
    "penalty": ['l1', 'l2', 'elasticnet', None],
    "solver": ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    "random_state": [20],
    "max_iter": [10000]
    }

model_GC = LogisticRegression()
grid = GridSearchCV(model_GC,
                    parametrs,
                    scoring='recall',
                    cv=6)

grid.fit(x_train, y_train)
grid.best_params_
```

Была обучена модель, используя найденные гиперпараметры и выведены её метрики.

```python
LR_model = LogisticRegression(penalty=None, solver='lbfgs', random_state=20, max_iter=10000)
LR_model.fit(x_train, y_train)

y_pred = LR_model.predict(x_valid)

acc = accuracy_score(y_valid, y_pred)
precision = precision_score(y_valid, y_pred)
recall = recall_score(y_valid, y_pred)
f1 = f1_score(y_valid, y_pred)
balanced_accuracy = balanced_accuracy_score(y_valid, y_pred)
print(classification_report(y_valid, y_pred))
print('balanced accuracy: ', balanced_accuracy)
print('accuracy: ', acc)
print('precision: ', precision)
print('recall: ', recall)
print('f1: ', f1)
```

* Balanced accuracy = 0.688 означает, что модель неплохо справляется с учетом дисбаланса.
* Accuracy = 0.747 — это достаточно хороший результат, который показывает, что модель правильно классифицирует 74.7% случаев.
* Precision = 0.571.Это вполне неплохой результат
* Recall = 0.545. Это означает, что модель пропускает значительную часть положительных примеров
* F1-Score = 0.558. Это значение говорит о том, что модель не идеальна, но все же показывает умеренный результат по сравнению с другими методами классификации.

### Случайный лес

Используя GridSearchCV, переберём все комбинации параметров модели случайного леса, чтобы найти лучшие значения гиперпараметров.

```python
parametrs = {
    "n_estimators": [2, 3, 4, 5, 10, 20, 50, 100],
    "criterion": ['gini', 'entropy', 'log_loss'],
    "max_depth": [2, 3, 4, 5, 6, 7, 8],
    "random_state": [20]
    }

model_GC = RandomForestClassifier()
grid = GridSearchCV(model_GC,
                    parametrs,
                    scoring='recall',
                    cv=6)

grid.fit(x_train, y_train)
grid.best_params_

```

Была обучена модель, используя найденные гиперпараметры и выведены её метрики.

```python
forest_model = RandomForestClassifier(n_estimators=50,
                                      criterion='entropy',
                                      max_depth=8,
                                      random_state=20)
forest_model.fit(x_train, y_train)

y_pred = forest_model.predict(x_valid)

acc = accuracy_score(y_valid, y_pred)
precision = precision_score(y_valid, y_pred)
recall = recall_score(y_valid, y_pred)
f1 = f1_score(y_valid, y_pred)
balanced_accuracy = balanced_accuracy_score(y_valid, y_pred)
print(classification_report(y_valid, y_pred))
print('balanced accuracy: ', balanced_accuracy)
print('accuracy: ', acc)
print('precision: ', precision)
print('recall: ', recall)
print('f1: ', f1)
```

* Balanced accuracy = 0.794 — это отличный результат, который указывает на то, что модель хорошо справляется с задачей, учитывая возможный дисбаланс классов.
* Accuracy = 0.84 — это очень высокий показатель, который говорит о том, что модель правильно классифицирует 84% всех случаев.
* Precision = 0.75 — это очень хороший результат. Это означает, что из всех случаев, когда модель предсказывает положительный класс, 75% действительно являются положительными.
* Recall = 0.682 говорит о том, что модель правильно нашла 68.2% всех положительных примеров. Это довольно хороший результат по сравнению с другими моделями.
* F1-Score = 0.714 — это хороший результат, который показывает, что модель довольно эффективно балансирует между точностью и полнотой.

### Общие результаты

Далее построены ROC-кривые для всех моделей на одном графике.

```python
fpr_knn, tpr_knn, _ = roc_curve(y_valid, neighbors_model.predict_proba(x_valid)[:, 1])
fpr_tree, tpr_tree, _ = roc_curve(y_valid, tree_model.predict_proba(x_valid)[:, 1])
fpr_logistic, tpr_logistic, _ = roc_curve(y_valid, LR_model.predict_proba(x_valid)[:, 1])
fpr_forest, tpr_forest, _ = roc_curve(y_valid, forest_model.predict_proba(x_valid)[:, 1])

plt.figure(figsize=(8, 6))
plt.plot(fpr_knn, tpr_knn, label='K-ближайших соседей')
plt.plot(fpr_tree, tpr_tree, label='Дерево решений')
plt.plot(fpr_logistic, tpr_logistic, label='Логистическая регрессия')
plt.plot(fpr_forest, tpr_forest, label='Случайный лес')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая для различных моделей')
plt.legend()
plt.grid()
plt.show()
```

Чем больше площадь под кривой, тем лучше модель справляется с задачей. По полученному графику можно сказать, что лучше всего справляются модели логической регрессии и случайного леса, а хуже всего - k-ближайших соседей.

Было оценено качество классификации с помощью метрик "Accuracy", "Precision", "Recall", "Balanced accuracy", "F1 score". Также была построена матрица неточностей.

```python
models = {
    'K-ближайших соседей': neighbors_model,
    'Логистическая регрессия': LR_model,
    'Дерево решений': tree_model,
    'Случайный лес': forest_model
}

fig, axes = plt.subplots(2, len(models), figsize=(16, 8))
metrix = []
i = 0
for name, model in models.items():
  y_pred = model.predict(x_valid)

  balanced_accuracy = balanced_accuracy_score(y_valid, y_pred)
  acc = accuracy_score(y_valid, y_pred)
  precision = precision_score(y_valid, y_pred)
  recall = recall_score(y_valid, y_pred)
  f1 = f1_score(y_valid, y_pred)

  dtc_matrix = confusion_matrix(y_valid, y_pred)
  sns.heatmap(dtc_matrix, annot=True, ax=axes[0][i], vmin=0)
  i += 1

  metrix.append( [name, balanced_accuracy, acc, precision, recall, f1])

for i, ax in enumerate(axes[1]):
  legend_elements = [Line2D([0], [0], color='w', label=f'{metrix[i][0]}'),
                     Line2D([0], [0], color='w', label=f'balanced accuracy: {str(round(metrix[i][1], 4))+" "*8}'),
                     Line2D([0], [0], color='w', label=f'accuracy: {str(round(metrix[i][2], 4))+" "*8}'),
                     Line2D([0], [0], color='w', label=f'precision: {str(round(metrix[i][3], 4))+" "*8}'),
                     Line2D([0], [0], color='w', label=f'recall: {str(round(metrix[i][4], 4))+" "*8}'),
                     Line2D([0], [0], color='w', label=f'f1: {str(round(metrix[i][5], 4))+" "*8}')
                     ]

  ax.legend(handles=legend_elements, loc='upper center')
  for spine in ax.spines.values():
      spine.set_alpha(0)
  ax.set_xticks([])
  ax.set_yticks([])
```

Исходные данные не очень сбалансированы, имеется численное преобладание 0-класса над 1-классом примерно в два раза. Из-за несбалансированности модель не может качественно обучится. Все они демонстрируют хорошую и среднюю точность от 61% (К-ближайших соседей) до 84% (Случайный лес) и средние показатели прочих метрик.

Целевая переменная - DEATH_EVENT. Модель случайного леса предсказывает её лучше всего. Таким образом, лучше использовать эту модель для получения наиболее вероятного прогноза случаев смерти.
